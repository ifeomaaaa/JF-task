```python
import pandas as pd 
df=pd.read_csv('imdb_top_1000.csv', encoding='utf8')
df['Runtime']=df['Runtime'].str.replace('min','') #remove column
df = df.dropna(subset=["Runtime"]) #dropnNaN because was affecting conversion to int
df.sort_values(by=['Runtime'], inplace=True)
df.tail(10)
```

```python
df['AVG_Meta_score'] = df.groupby('Runtime').Meta_score.transform('mean')
df.head(8)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(70,30))
ax.plot(df["Runtime"], df["AVG_Meta_score"], color="red")
plt.title("Average Meta Score per Runtime") 
plt.xlabel("Runtime")
plt.ylabel("Average Meta Score") 
plt.show() 
```
