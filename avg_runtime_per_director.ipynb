```python
import pandas as pd 
df=pd.read_csv('imdb_top_1000.csv', encoding='utf8')
df['Gross']=df['Gross'].str.replace(',','') #remove column
df['Runtime']=df['Runtime'].str.replace('min','') #remove min from runtime
df = df.dropna(subset=["Gross"]) #dropnNaN because was affecting conversion to int
df = df.dropna(subset=["Runtime"]) #dropnNaN because was affecting conversion to int
df['Gross'] = df['Gross'].astype(int) #convert to int
df['Runtime'] = df['Runtime'].astype(int) #convert to int
df.sort_values(["Released_Year"], 
                    axis=0,
                    ascending=[True], 
                    inplace=True)
df.head(3)
```

```python
df1=df.groupby("Director").size()
df1.shape
```

```python
df['AVG_Runtime'] = df.groupby('Director').Runtime.transform('mean')
df.head(8)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(500,150))
ax.bar(df["Director"], df["AVG_Runtime"], color="red")
plt.title("Average Runtime per Director") 
plt.xlabel("Director")
plt.ylabel("Average Runtime")
plt.xticks(rotation = 45) # Rotates X-Axis Ticks by 45-degrees
plt.show() 

```
